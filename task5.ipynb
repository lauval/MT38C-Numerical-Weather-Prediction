## The Analytical Solution

This time we're looking to solve our equations on a 2D grid. The same equations and conditions apply. So recall that, for non-divergent flow, under geostrophic balance, the vorticity equation is given by:
$$ \xi = \frac{2A}{L^2_y} \frac{\phi_0}{f_0} \tanh{\left(\frac{y-y_0}{L_y}\right)} \left[1 - \tanh^2 (\frac{y-y_0}{L_y})\right]$$

We also have the horizontal velocity, u, given by:
$$ u = \frac{A}{L_y} \frac{\phi_0}{f_0} \left[1 - \tanh^2 (\frac{y-y_0}{L_y})\right]$$

And we have the stream function given by:
$$ \Psi = \frac{\Phi_0}{f_0} \left[ 1 - A \tanh{\left(\frac{y-y_0}{L_y}\right)}\right] $$

Again, we want to calculate and plot $\Psi(y)$, $u(y)$ and $\xi(y)$ for the following parameters:

- $\phi_0 = 10^{5}$ metres squared per second squared
- $A = 10^{-4}$ per second
- $L_y = 2.5 \times 10^{5}$ metres
- $A = 5 \times 10^{-3}$
- $y_0 = 5 \times 10^5$ metres

```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
from scipy.linalg import solve

plt.style.use('science')
```

## The Analytical Solution

First we create a meshgrid:

```python
# define a common grid spacing
grid_space = 1e5

# adding 1e5 to both limits so that np.arrange() includes the upper limit
upper_x_limit = 2.5e7 + grid_space
upper_y_limit = 5e6 + grid_space

x = np.arange(0, upper_x_limit, grid_space)
y = np.arange(0, upper_y_limit, grid_space)

# define an X array where :
Y, _ = np.meshgrid(y, x)
```

Visualise arrays. We can see that values increase as the column number increases for the X array. Y values increase with increasing row number.

```python
print("Y: \n", Y)
```

Then define all constants:

```python
phi_0 = 1e5
f_0 = 1e-4
L_y = 2.5e5
A = 5e-3

y_0 = 2.5e06
```

We also define a simple tanh function to simplify the syntax.

```python
def tanh(x):
    return np.tanh(x)
```

Now we calculate the stream function, u and vorticity:

```python
vorticity = (2*A/(L_y**2)) * (phi_0/f_0) * tanh((Y-y_0)/L_y) * (1 - ((tanh((Y-y_0)/L_y))**2))

u = A/L_y * phi_0/f_0 * (1 - ((tanh((Y-y_0)/L_y))**2))

streamfunction = phi_0/f_0 * (1 - (A * (tanh((Y-y_0)/L_y))))
```

And we visualise the results:

Streamfunction, $\psi$

```python
fig, ax = plt.subplots(1,2, figsize = (4,2), dpi=250)
ax[0].contourf(streamfunction)
ax[1].plot(streamfunction)
fig.suptitle(r'$\psi$')
fig.subplots_adjust(top=0.8, hspace=2, wspace=0.5)
plt.show()
```

Velocity field, $u$

```python
fig, ax = plt.subplots(1,2, figsize = (4,2), dpi=250)
ax[0].contourf(u)
ax[1].plot(u)
fig.suptitle(r'$u$')
fig.subplots_adjust(top=0.85, hspace=2)
plt.show()
```

Vorticity, $\xi$

```python
fig, ax = plt.subplots(1,2, figsize = (4,2), dpi=250)
ax[0].contourf(vorticity)
ax[1].plot(vorticity)
fig.suptitle(r'$\xi$')
fig.subplots_adjust(top=0.8, hspace=2)
plt.show()
```

## Finite Difference Forms

We also have these forms of the same equations:

$$ u = -\frac{\partial \Psi}{\partial y} $$

$$ \xi = \frac{\partial^2 \Psi}{\partial y^2} $$

For the velocity, we can use a central difference scheme:
$$ u = \frac{\Psi_{j, i+1} - \Psi_{j, i-1}}{2 \Delta y} $$

This gives us a second order accurate scheme, however, we must account for the boundary conditions. We can use a forward difference scheme for the first point: 
$$ u_0 = \frac{\Psi_{1} - \Psi_{0}}{\Delta y} $$

And a backward difference scheme for the last point:
$$ u_i = \frac{\Psi_{j, i} - \Psi_{j, i-1}}{\Delta y} $$

Now we can code up a solution for the velocity field using these schemes:

### Flow Speed

```python
def generate_velocity(streamfunction, y):
    N = np.shape(Y)[1] # for index slicing in the centred difference loop
    dy = y[0][1]-y[0][0] # extract grid space step from y array
    
    # implement centred difference scheme for 2D array    
    u = -1 * ((streamfunction[:,2:N] - streamfunction[:,0:(N-2)]) / (2 * dy))

    return u
```

And now we can visualise the results:

```python
numerical_velocity = generate_velocity(streamfunction, Y)

fig, ax = plt.subplots(2,2, figsize=(8,3), dpi=200)
ax[0,0].contourf(numerical_velocity)
ax[0,1].plot(numerical_velocity)

ax[1,0].contourf(u)
ax[1,1].plot(u)

ax[0,0].set_title('Numerical Velocity')
ax[1,0].set_title('Analytical Velocity')

fig.subplots_adjust(hspace=0.5)

fig.suptitle('Comaprison of Numerical Velocity and Analytical Velocity on a 2D Grid', y=1.05)
```

### Vorticity

We can approximate the second derivative as follows:

$$ \frac{\partial^2 \Psi}{\partial y^2} \approx \frac{\Psi_{i+1} - 2\Psi_i + \Psi_{i-1}}{\Delta y^2} $$

Which we code as follows:

```python
def generate_vorticity(streamfunction, y):
    N = np.shape(Y)[1] # for index slicing in the centred difference loop
    dy = y[0][1]-y[0][0] # extract grid space step from y array
    
    # implement centred difference scheme for 2D array    
    u = (streamfunction[:,2:N] - (2 * streamfunction[:,1:N-1]) + streamfunction[:,0:(N-2)]) / (dy**2)

    return u
```

Visualise results

```python
numerical_vorticity = generate_vorticity(streamfunction, Y)

fig, ax = plt.subplots(2,2, figsize=(8,3), dpi=200)
ax[0,0].contourf(numerical_vorticity)
ax[0,1].plot(numerical_vorticity)

ax[1,0].contourf(vorticity)
ax[1,1].plot(vorticity)

ax[0,0].set_title('Numerical Vorticity')
ax[1,0].set_title('Analytical Vorticity')

fig.subplots_adjust(hspace=0.5)

fig.suptitle('Comaprison of Numerical Vorticity and Analytical Vorticity on a 2D Grid', y=1.05)
```
